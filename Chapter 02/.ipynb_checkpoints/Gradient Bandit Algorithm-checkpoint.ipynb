Here we introduce new approach to give numerical preference($H_t(a)$) for some action. The larger the preference is the more that action is taken. Here actions are taken relative to the other actions.

$$
\begin{gather}
    Pr\{A_t=a\} = \text{Softmax}\big(H_t(a)\big) = \frac{e^{H_t(a)}}{\sum_{b=1}^{k}e^{H_t{b}}} = \pi_t(a)
\end{gather}
$$

Initially preferences are zero: $H_1(a) = 0$, so all actions have equal probability of being selected.
There is a natural learning algorithm for this setting based on the idea of stochastic gradient ascent. On each step after selecting action $A_t$ and receiving reward $R_t$ the action preferences are updated by

$$
\begin{gather}
    H_{t + 1}(A_t) = H_t(A_t) + \alpha(R_t - \bar{R_t})(1 - \pi(A_t))\\
    H_{t + 1}(a) = H_t(a) - \alpha(R_t - \bar{R_t})\pi_t(a) & \text{for all $a \ne A_t$}
\end{gather}
$$

$\alpha \gt 0$ : step-size parameter <br>
$\bar{R_t} \in \mathbb{R}$ : average of all the rewards up through and including time $t$

### The Bandit Gradient Algorithm as stochastic gradient ascent
We should define rewards expected to be observed at step $t$:
$$
\begin{gather}
    \mathbb{E}[R_t] = \sum_x \pi_t(x)q_*(x)
\end{gather}
$$
where $x$ is some action. Our objective here is to maximize the expected reward $\mathbb{E}[R_t]$. <br>
As $\pi_t(x)$ shows the probability of the action $x$ being an optimal action to take, then our objective would be to make the optimal action's probability higher so the term $\pi_t(x)q_*(x)$ give us maximum reward. <br>
To optimize this
$$
\begin{gather}
    H_{t+1}(a) = H_t(a) + \alpha\frac{\partial\mathbb{E}[R_t]}{\partial H_t(a)}
\end{gather}
$$
the update above is moving toward maximum. However it is not possible to implement gradient ascent exactly like this because by assumption we do not know the $q_*(x)$.
#### Take a look at the performance of the gradient
$$
\begin{gather}
    \frac{\partial\mathbb{E}[R_t]}{\partial H_t(a)} = \frac{\partial}{\partial H_t(a)} \bigg[\sum_x \pi_t(x)q_*(x)\bigg]\\
    \text{Since gradient of sums is sum of gradients}\\
    = \sum_x q_*(x)\frac{\partial\pi_t(x)}{\partial H_t(a)}\\
    \text{Add the baseline term}\\
    = \sum_x \big(q_*(x) - B_t\big)\frac{\partial\pi_t(x)}{\partial H_t(a)}
\end{gather}
$$
$B_t$, called _baseline_, can be any scalar that does not depend on $x$. We can include a baseline here without changing the equality becuase the gradient sums to $0$ overall actions:
$$
\begin{gather}
    \text{As we know $\sum_x \pi_t(x) = 1$}\\
    \sum_x \frac{\partial\pi_t(x)}{\partial H_t(a)}\\
    \text{Sum of derivatives is derivative of sums}\\
    = \frac{\partial}{\partial H_t(a)}\sum_x\pi_t(x)\\
    = \frac{\partial}{\partial H_t(a)}1 = 0
\end{gather}
$$

As $H_t(a)$ is changed, some actions' probabilities go up and some go down, but the sum of the changes must be zero because the sum of the probabilities is always one.

Now we multiply and divide each term of the sum by $\pi_t(x)$ to make it in the form of expectation:
$$
\begin{gather}
    \frac{\partial\mathbb{E}(R_t)}{\partial H_t(a)} = \sum_x \pi_t(x)\big(q_*(x) - B_t\big)\frac{\partial\pi_t(x)}{\partial H_t(a)}/\pi_t(x)
\end{gather}
$$
Now the equation is in the form of expectation, summing over all possible values $x$ of the random variable $A_t$, then multiplying by the probability of taking those values. Thus:
$$
\begin{gather}
    = \mathbb{E}\bigg[\big(q_*(A_t) - B_t\big)\frac{\partial\pi_t(A_t)}{\partial H_t(a)}/\pi_t(A_t)\bigg]\\
    = \mathbb{E}\bigg[\big(R_t - \bar{R_t}\big)\frac{\partial\pi_t(A_t)}{\partial H_t(a)}/\pi_t(A_t)\bigg]
\end{gather}
$$
We call the baseline $\bar{R_t}$ and substitute $R_t$ for $q_*(A_t)$ which is permitted because $\mathbb{E}[R_t|A_t] = q_*(A_t)$. Shortly we will establish that $\frac{\partial\pi_t(x)}{\partial H_t(a)}=\pi_t(x)\big(\mathbb{1}_{a=x}-\pi_t(a)\big)$ where $\mathbb{1}_{a=x}$ if defined to be $1$ if $a = x$, else 0.
Thus we have:
$$
\begin{gather}
    = \mathbb{E}\big[(R_t - \bar{R_t})\pi_t(A_t)\big(\mathbb{1}_{a=x}-\pi_t(a)\big)/\pi_t(A_t)\big]\\
    = \mathbb{E}\big[(R_t - \bar{R_t})\big(\mathbb{1}_{a=x}-\pi_t(a)\big)\big]
\end{gather}
$$

Recall that our plan has been to write the performance gradient as an expectation of something that we can sample on each step, as we have just done, and then update on each step in proportion to the sample. Substituting a sample of expectation above for the performance gradient yields:

$$
\begin{gather}
    H_{t+1}(a) = H_t(a) + \alpha\frac{\partial\mathbb{E}[R_t]}{\partial H_t(a)}\\
     = H_t(a) + \alpha(R_t - \bar{R_t})\big(\mathbb{1}_{a=x}-\pi_t(a)\big), & \text{for all $a$}
\end{gather}
$$
<br>
#### Showing $\frac{\partial\pi_t(x)}{\partial H_t(a)} = \pi_t(x)\big(\mathbb{1}_{a=x}-\pi_t(a)\big)$
$$
\begin{gather}
    \frac{\partial\pi_t(x)}{\partial H_t(a)} = \frac{\partial}{\partial H_t(a)}\pi_t(x)\\
    = \frac{\partial}{\partial H_t(a)}\bigg[\frac{e^{H_t(x)}}{\sum_{y=1}^k e^{H_t(y)}}\bigg]\\
    = \frac{\frac{\partial e^{H_t(x)}}{\partial H_t(a)}\sum_{y=1}^k e^{H_t(y)} - e^{H_t(x)}\frac{\partial\sum_{y=1}^k e^{H_t(y)}}{\partial H_t(a)}}{\bigg(\sum_{y=1}^k e^{H_t(y)}\bigg)^2} & \text{(by the quotient rule)}\\
    = \frac{\mathbb{1}_{a=x}\sum_{y=1}^k e^{H_t(y)} - e^{H_t(x)}e^{H_t(a)}}{\bigg(\sum_{y=1}^k e^{H_t(y)}\bigg)^2} & \text{(because $\frac{\partial e^x}{\partial x} = e^x$)}\\
    = \frac{\mathbb{1}_{a=x}e^{H_t(y)}}{\sum_{y=1}^{k}e^{H_t(y)}} - \frac{e^{H_t(x)}e^{H_t(a)}}{\bigg(\sum_{y=1}^{k}e^{H_t(y)}\bigg)^2}\\
    = \mathbb{1}_{a=x}\pi_t(x) - \pi_t(x)\pi_t(a)\\
    = \pi_t(x)\big(\mathbb{1}_{a=x}-\pi_t(a)\big)
\end{gather}
$$

_We have just shown that the expected update of the gradient bandit algorithm
is equal to the gradient of expected reward, and thus that the algorithm is an
instance of stochastic gradient ascent. This assures us that the algorithm has robust
convergence properties._

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Define pi_t(a)
def softmax(H):
    """
    args:
        H:            All the numerical preferences for each action

    returns:
        pi_t:         Probability of each action for being the optimal action
    """
    return np.exp(H) / np.sum(np.exp(H))
````

````python
def gradient_bandit(n, alpha, steps, runs, stationary=True, verbose=1, **kwargs):
    """
    args:
        n:           Number of arms in k-armed bandit task
        alpha:       Step-size parameter
        steps:       Number of time-steps
        runs:        Number of independent runs
        stationary:  A boolean that shows if the task stationary
        verbose:     Verbose 0 does not print progress, 1 shows how many runs are completed, 2 shows last avg_reward

    returns:
        mean_rewards at each step
        mean of ratio of optimal actions selected
    """

    # Define the numertical preference for each action
    H = np.zeros(n)
    # Define the baseline R_bar (Average rewards observed including time-step t)
    R_bar = 0
    # Define avg_reward: mean rewards observed over all runs
    avg_rewards = np.zeros(steps)
    # Define avg_optimal: mean ratio of optimal ations selected over all runs
    avg_optimal = np.zeros(steps)
    # mean and variance for probability distribution of q_star
    q_star_mean = kwargs['q_star_mean'] if 'q_star_mean' in kwargs else 0
    q_star_var = kwargs['q_star_var'] if 'q_star_var' in kwargs else 1
    
    for run in range(runs):
        # Reset H, R_bar, and q_star for each run
        q_star = np.random.normal(loc=q_star_mean, scale=q_star_var, size=n)
        H.fill(0)
        R_bar = 0
        
        for t in range(steps):
            # Calculate pi_t
            pi = softmax(H)

            # Take an action, which each action having the probability pi_t(a)
            A = np.random.choice(np.arange(n), p=pi)

            # Observe a reward
            R = q_star[A] + np.random.randn()
    
            # Update R_bar
            R_bar = R_bar + (R - R_bar) / (t + 1)
            
            # Update the numerical preference
            predicate = np.zeros(n)
            predicate[A] = 1
            H = H + alpha * (R - R_bar) * (predicate - pi)
            # Average over rewards observed
            avg_rewards[t] += R / runs
    
            # Average over the ratios of which true actions are selected
            avg_optimal[t] += (1 if A == np.argmax(q_star) else 0) / runs
            
            # Update q_star if not stationary
            if not stationary: q_star = q_star + np.random.randn(n) * 0.01

        if verbose == 1:
            print(f'\rRun {run + 1:>4} out of {runs:>4}', end='')
        elif verbose == 2:
            print(f'\rRun {run + 1:>4} out of {runs:>4} avg_reward: {avg_rewards[steps - 1]:0.3e} {" ":>3}', end='')

    return avg_rewards, avg_optimal
````

````python
def plot(ax1, ax2, avg_rewards, avg_optimal, label):
    line, = ax1.plot(avg_rewards, label=label)
    ax1.legend(handles=ax1.lines)
    ax1.set_xlabel('time-steps $t$')
    ax1.set_ylabel('average rewards')
    ax1.set_title('Average rewards observed at each time-step');
    
    line, = ax2.plot(avg_optimal * 100, label=label)
    ax2.legend(handles=ax2.lines, loc='best')
    ax2.set_xlabel('time-steps $t$')
    ax2.set_ylabel('%optimal action taken')
    ax2.set_title('Optimal action taken at each time-step');
    ax2.set_ylim([0, 100]);
    ax2.set_yticks(range(0, 101, 20));
    ax2.set_yticklabels(['%' + str(x) for x in range(0, 101, 20)]);
````

````python
n = 10                             # Number of arms in k-armed bandit task
alpha = 0.1                        # Step-size Parameter
steps = 1000                       # Number of time-steps
runs = 1000                        # Number of independent runs

rewards, optimal = gradient_bandit(n, alpha, steps, runs, verbose=2)
rewards_m4, optimal_m4 = gradient_bandit(n, alpha, steps, runs, verbose=2, q_star_mean=4.0)
rewards_nonstationary, optimal_nonstationary = gradient_bandit(n, alpha, steps, runs, stationary=False, verbose=2)

fig = plt.figure(figsize=(8, 11))
ax1 = plt.subplot(2, 1, 1)
ax2 = plt.subplot(2, 1, 2)
plot(ax1, ax2, rewards, optimal, label='stationary')
plot(ax1, ax2, rewards_m4, optimal_m4, label='stationary q_star_mean=4.0')
plot(ax1, ax2, rewards_nonstationary, optimal_nonstationary, label='nonstationary')
````

### Notes
$\bar{R}_t$ is updated as follows:
$$
\bar{R}_{t+1} = \bar{R}_t + \frac{1}{t}(R - \bar{R}_t)
$$

Also, Gradient Bandit algorithm uses $\pi_t$ to explore instead of exploring with probability of $\epsilon$ by choosing an action randomly. Instead of choosing each action randomly, an action is selected with the probability $\pi_t(A_t)$. This means that actions are selected from a probability distribution with each action having probability $\pi_t(a)$.

This could be implemented as follows:
```python
    A = np.random.choice(np.arange(n), p=pi)
```
where ```n``` is the number of arms and ```pi``` is $\pi_t$.
