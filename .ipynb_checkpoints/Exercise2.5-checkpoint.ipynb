Design and conduct an experiment to demonstrate the difficulties that sample-average methods have for nonstationary problems. Use modified version of 10-arm testbed in which all $q_*(a)$ start out equal and then take independant random walks (say by adding a normally distributed increment with mean 0 and standard deviation 0.01 to all the $q_*(a)$ on each step). Prepare plots like Figure 2.2 for an action-value method using sample-averages, incrementally computed and another action-value method with constant step-size parameter, $\alpha=0.1$. Use $\epsilon = 0.1$ and longer runs, say of 10,000 steps.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10                          # Number of Arms
steps = 10000                   # Number of steps denoted by t
runs = 2000                     # Number of runs to run each step by and average over
epsilon = 0.1                   # Probability that the agents explores
alpha = 0.1                     # Constant step-size

# initialize true action-values
q_star = np.random.randn(n)

# initialize estimates and counts for sample-average method
Q1 = np.zeros(n)
N1 = np.zeros(n)

# initialize estimates for constant step-size method
Q2 = np.zeros(n)

# initialize average rewards for each method
avg_rewards1 = np.zeros(steps)
avg_rewards2 = np.zeros(steps)

for run in range(runs):
    # We run each step by some number of bandits to smooth the curves
    q_star = np.random.randn(n)
    Q1.fill(0)
    N1.fill(0)
    Q2.fill(0)
    for t in range(steps):
        # Choose action for sample-average method
        if np.random.rand() < epsilon:
            A1 = np.random.choice(n)
        else:
            A1 = np.argmax(Q1)

        # Choose action for constant step-size method
        if np.random.rand() < epsilon:
            A2 = np.random.choice(n)
        else:
            A2 = np.argmax(Q2)

        # Generate reward based on true action-values
        R1 = q_star[A1] + np.random.randn()
        R2 = q_star[A2] + np.random.randn()

        # Update counts and estimate-values for sample-average method
        N1[A1] += 1
        Q1[A1] += (R1 - Q1[A1]) / N1[A1]

        # update estimate-values for constant step-size method
        Q2[A2] += alpha * (R2 - Q2[A2])

        # update q_star for this bandit at this time-step
        q_star += 0.01 * np.random.randn(n)

        # add rewards to sum to be averaged
        avg_rewards1[t] += R1 / runs
        avg_rewards2[t] += R2 / runs

    # print progress
    print(f'\rCompleted step {run + 1:0>4} out of {runs:0>4}', end='')
```

```python
# Plot results
plt.plot(avg_rewards1, label="Sample-average")
plt.plot(avg_rewards2, label="Constant step-size")
plt.xlabel("Steps")
plt.ylabel("Average reward")
plt.legend()
plt.show()
```

## Explaination
The problem is nonstationary meaning that $q_*(a)$ changes from time-step to time-step; so for each time step we let 1000 bandits in a decision making with two methods:

- Sample-average method
- Constant step-size
  
In each run we take an action $A$ greedily corresponding to their estimate-values($Q$), where $Q_1$ denotes estimate-values for sample-average method and $Q_2$ denotes estimate-values for constant step-size method, also $A_1$ is the action corresponding to estimate-value for sample-average method and $A_2$ is the action corresponding to constant step-size method; alternatively with probability of $\epsilon$, they take a random action.
Each of the actions $A_1$ and $A_2$ observe a reward which is a function of $q_*(a)$ (in this case).

### Updating estimate-value for sample-average method
$Q_t$ keeps being the average over all rewards seen for some action $a$
$$
\begin{gather}
    Q_{t+1}(a) = Q_t + \frac{1}{n}\big[R_t - Q_t\big]
\end{gather}
$$

### Update estimate-value for constant step-size method (exponential recency-weighted average)
$Q_t$ is the weighted average of the rewards in which recent rewards are given more weight then older ones. ($\alpha$ is a constant)
$$
\begin{gather}
    Q_{t+1} = Q_t + \alpha\big[R_t - Q_t\big]\\
    Q_{t+1} = Q_t + \alpha R_t - \alpha Q_t\\
    Q_{t+1} = \alpha R_t + (1 - \alpha)Q_t\\
    Q_{t+1} = \alpha R_t + (1 - \alpha)\big[\alpha R_{t-1} + (1 - \alpha)Q_{t-1}\big]\\
    Q_{t+1} = \alpha R_t + (1 - \alpha)\alpha R_{t-1} + (1 - \alpha)^2Q_{t-1}\\
    Q_{t+1} = \alpha R_t + (1 - \alpha)\alpha R_{t-1} + (1 - \alpha)^2\alpha R_{t-2} + (1 - \alpha)^3Q_{t-2}\\
    Q_{t+1} = \alpha R_t + (1 - \alpha)\alpha R_{t-1} + (1 - \alpha)^2\alpha R_{t-2} + \dots + (1 - \alpha)^{t}Q_1\\
    Q_{t+1} = (1 - \alpha)^{t}Q_1 + \sum_{i=1}^{t} \alpha(1-\alpha)^{t-i}R_i
\end{gather}
$$

Since $0 \lt \alpha \le 1$ then $0 \le 1 - \alpha \lt 1$. This shows that if $t - i$ is large then the weight given to a reward is less; in other word recent rewards, the ones who are close to the time-step $t$ get more weight:
- At time-step $t$: the weight for $R_t$ is $\alpha(1-\alpha)^{t-t}$ which is $\alpha$
- At time-step $t-1$: the weight for $R_{t-1}$ is $\alpha(1-\alpha)^{t - t + 1}$ which is $\alpha(1-\alpha)$ which is less than $\alpha$
- At time-step $1$: the weight for $R_1$ is $\alpha(1-\alpha)^{t-1}$
